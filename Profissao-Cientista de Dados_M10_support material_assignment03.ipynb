````python
#Importando bibliotecas necessárias. 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
#Carregando o Dataframe. 
prev_renda = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
prev_renda.head()
````

````python
# Verificando quais colunas devem ser usadas. 
prev_renda.info
````

````python
#Escolhendo variáveis qualitativas. 
var_quali = ["sexo", "posse_de_veiculo", "posse_de_imovel", "tipo_renda", "educacao", "estado_civil", "tipo_residencia","mau"]
````

````python
#Tranformando a coluna “data_ref” para o tipo datetime.
prev_renda ['data_ref']= pd.to_datetime (prev_renda['data_ref'])
````

````python
#Plotando os gráficos para cada variável qualitativa ao longo do tempo.
for var in var_quali:
    plt.figure(figsize = (12,6))
    plt.title(f'Distribuição da variável {var} no tempo')
    sns.countplot(x= 'data_ref', hue = var, data= prev_renda)
    plt.xticks(rotation=45)
    plt.xlabel('Data de Referência')
    plt.ylabel('Contagem')
    plt.legend(loc = 'upper left')
   # plot.show()
````

# Resposta:
- Só foi possível notar diferença considerável ao longo do tempo na variável educação.
- Onde a partir do dia 01/09/2015 a diferença entre o nível secundário e o superior completo reduziu consideravelmente. Com a redução no nível secundário e aumento no superior completo. 


2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
plt.figure(figsize=(20,20))
for i, var in enumerate(var_quali, 1):
    plt.subplot(3, 3, i)
    sns.pointplot(x='data_ref', y='renda', hue=var, data=prev_renda)
    plt.title(f"Médias de Renda por {var}")
    plt.xlabel("Data de Referência")
    plt.ylabel("Média de Renda")
    plt.xticks(rotation=45)
    plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
````

# Resposta:

- Analisando as medias de renda por sexo é possível notar uma grande diferença entre o sexo masculino e o feminino que se manteve por todo o período observado.
- Durante a avaliação da variável posse de veículo podemos notar uma diferença inferior a variável anterior, porem ainda presente, com uma renda maior para os que possuem veículo. Essa diferença permanece pelo período estudado.
- Na variável posse de imóvel mais uma vez a diferença diminui, mas ainda pode ser notada com uma vantagem para os que possuem imóveis.
- Com a variável tipo de renda é possível notar uma queda brusca nos bolsistas entre as datas: 02/2015 e 03/2015. Também é perceptível uma boa diferença entre os pensionistas e os servidores públicos, com os assalariados e empresários em uma posição intermediária.
- Na analise da variável educação como esperado as rendas mais altas são de pessoas com maior escolaridade, os pós-graduados, com pouca diferença entre o restante.
- Com a avaliação da variável estado civil podemos notar que os viúvos apresentam uma renda menor e com oscilações durante todo o período nas outras categorias.
- É possível perceber que na variável tipo de residência as categorias se mantem próximas umas das outras a maior parte do tempo.
- Podemos notar que na variável mau os clientes classificados com bons tem uma renda um pouco maior e com menos oscilações.
- Concluindo que pela falta de estabilidade as variáveis: sexo e posse de veículo não são ideais para serem usadas em uma modelagem, mas as outras podem sim ser usadas.
